```python
# Checking that encoding reversal works.

rev = le.inverse_transform(Labels)
rev = pd.Series(rev)
rev.unique()
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(Features, Labels, test_size=.2)

features_train.shape, features_test.shape, labels_train.shape, labels_test.shape
```

**Building ANN**<br>

Once the model is created,<br>

config the model with losses and metrics with model.compile()<br>

train the model with model.fit()<br>

use the model to do prediction with model.predict()<br>

```python
import tensorflow as tf
```

```python

ANN = tf.keras.models.Sequential([
    
    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(15, activation="softmax")
])
```

```python
#callback that stops training when minimum value of validation loss function is reached
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5, mode='min')
```

```python
ANN.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = ANN.fit(features_train,labels_train,
          epochs=50, 
          batch_size=256,
          callbacks=[early_stopping])
```

autoencoder can predict 14 attacks and normal traffic with 99% accuracy

```python
preds = ANN.predict(features_test)
```

```python
labels_test_predictions = np.argmax(preds, axis=1)
```

```python
labels_test_predictions[:], labels_test[:]
```

**Accuracy** :This is the ratio of instances that have been properly identified out of the overall number of instances. <br>

Accuracy = TP + TN / TP + TN + FP + FN. <br><br>

**Precision**:This is the ratio of instances that have been correctly predicted as attacks out of all the samples that have been predicted as attacks. <br>

Precision = TP / TP + FP<br><br>

**Recall**: This is the ratio of instances that have been correctly classified as attacks out of all samples of actual attacks.<br> 

Detection Rate = TP / TP + FN<br><br>

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score

def print_statuses(labels, predictions):
  print("Accuracy = {}".format(accuracy_score(labels, predictions)))
  print("Precision = {}".format(precision_score(labels, predictions, average='macro')))
  print("Recall = {}".format(recall_score(labels, predictions, average='macro')))
```

```python
print_statuses((labels_test).astype(int),(labels_test_predictions).astype(int))
```
