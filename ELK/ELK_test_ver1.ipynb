```python
from elasticsearch import helpers
es_client = Elasticsearch(http_compress=True)
def doc_generator(df):
    df_iter = df.iterrows()
    for index, document in df_iter:
        yield {
                "_index": 'test_index_1',
                "_type": "_doc",
                "_id" : f"{document['ID']}",
                "_source": filterKeys(document),
            }
    #raise StopIteration
helpers.bulk(es_client, doc_generator(Elastic_df))
```
